```python
# hmmmmm it worked without error - but isn't correct

# apparentally there is a chardet package

import chardet


def detect_encoding(filename: str) -> str:
    """Return the most probable characte encoding for a file"""

    with open(filename, "rb") as f:
        raw_data = f.read()
        result = chardet.detect(raw_data)
        return result["encoding"]


detect_encoding("data/directors.csv")
```

```python
# looking at json

!cat data/pokedex.json
```

```python
# lets read it in

pokedex = pl.read_json("data/pokedex.json")
pokedex

# all one value because there is only one key: pokemon - booo
```

```python
# we can use explore which turns every item in a list into a new row
# and then use unnest which will turn every key into a new column
# very cool - need to see if R has something like this...

# I wanna see that step by step
# maybe practice caching....

plosion = pokedex.explode("pokemon").lazy()

# pokedex.explode("pokemon")

plosion.collect()
```

```python
unnest = plosion.unnest("pokemon").lazy()

unnest.collect()

# hey looks like a pretty normal dataframe, sweet
```

```python
unnest.select(
    "id",
    "name",
    "type",
    "height",
    "weight"
).collect()

# hey look at that...
```

```python
# ndjson
# each line in an ndjson is a valid json object, the whole document is not a valid json
# object because in between each row is a newline which is not part of the json spec

!cat data/wikimedia.ndjson

# cool
```

```python
from json import loads
from pprint import pprint

with open("data/wikimedia.ndjson")as f:
    pprint(loads(f.readline()))
```

```python
# notice that length, meta and revision are still nested we can unnest those

wikimedia = pl.read_ndjson("data/wikimedia.ndjson")
wikimedia
```

```python
(
    # rename here otherwise the unnesting fails due to dupe names
    wikimedia.rename({"id": "edit_id"})
    .unnest("meta")
    .select("timestamp", "title", "user", "comment")
)
```
